# Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

# Data Collection & Analysis

```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning/Medical Insurance Cost Prediction/insurance.csv')
```

# **Categorical Features:**

# sex
male	---1

female	-0

```python
dataset['sex']=dataset['sex'].map({'male':1,'female':0})
```

```python
dataset.head()
```

# region
southeast	--1

southwest --2

northwest	--3

northeast	--4

```python
dataset['region']=dataset['region'].map({'southeast':1,'southwest':2,'northwest':3,'northeast':4})
```

# smoker
no	--0

yes --1

```python
dataset['smoker']=dataset['smoker'].map({'no':0,'yes':1})
```

```python
dataset.head()
```

```python
dataset.shape
```

# Splitting the Features and Target

```python
x = dataset.drop(['charges'],axis=1)
```

```python
y= dataset['charges']
```

#Splitting the data into Training data & Testing Data

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

#Model training

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation

```python
#train data
train_data_prediction = model.predict(x_train)
r2_train = metrics.r2_score(y_train,train_data_prediction)
print('R squared value on train data : ',r2_train)
```

```python
#test data
test_data_prediction = model.predict(x_test)
r2_test = metrics.r2_score(y_test,test_data_prediction)
print('R squared value on test data : ',r2_train)
```

# Building a Predictive System
